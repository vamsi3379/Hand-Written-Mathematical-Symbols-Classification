```python
data_train = np.load('/content/drive/MyDrive/Project/data_train.npy').T
labels_train = np.load('/content/drive/MyDrive/Project/t_train.npy')

print(data_train.shape, labels_train.shape)
```

```python
# Labels Integer Encoding

labels_names = ['x','square root','plus sign','negative sign','equal','percent','partial','product','pi','summation']
```

```python
dimension=200
arraym=[]
for x in data_train:
    arraym.append(cv2.resize(x.reshape(300,300),(dimension,dimension)).reshape(dimension*dimension))
resc_X=np.array(arraym)

X_train_data, X_test, t_train_data, t_test = train_test_split(resc_X, labels_train, shuffle=True, stratify=labels_train, test_size=0.15)
np.save('X_test_data.npy', X_test)
np.save('t_test_data.npy', t_test)
X_train, X_val, t_train, t_val = train_test_split(X_train_data, t_train_data,shuffle=True, stratify=t_train_data,test_size=0.2)

print(X_train.shape, t_train.shape, X_val.shape, t_val.shape, X_test.shape, t_test.shape)


#Reshaping Xtrain and validation training
rgb_xtrain = np.array([ cv2.cvtColor(x, cv2.COLOR_GRAY2RGB) for x in X_train])
X_train_rs = tf.constant(rgb_xtrain.reshape((rgb_xtrain.shape[0],dimension,dimension, 3)), dtype=tf.float32)
rgb_val = np.array([ cv2.cvtColor(x, cv2.COLOR_GRAY2RGB) for x in X_val])
X_val_rs = tf.constant(rgb_val.reshape((rgb_val.shape[0],dimension,dimension, 3)), dtype=tf.float32)
```

```python
init_model = ResNet50(input_shape=(dimension,dimension,3), include_top=False, weights="imagenet")
for i in init_model.layers[:-5]:
    i.trainable=False
total_noOf_classes = len(labels_names)
tf.random.set_seed(21)
```

```python
cnn_model = tf.keras.models.Sequential([
    init_model,
    GlobalAveragePooling2D(),    
    Dense(500,  kernel_initializer='lecun_normal', use_bias = False), 
    keras.layers.BatchNormalization(),
    keras.layers.Activation('relu'),
    Dropout(0.8),
    Dense(250,  kernel_initializer='lecun_normal', use_bias = False), 
    keras.layers.BatchNormalization(),
    keras.layers.Activation('relu'),
    Dropout(0.6),
    keras.layers.Dense(100,  kernel_initializer='lecun_normal', use_bias = False), 
    keras.layers.BatchNormalization(),
    keras.layers.Activation('relu'),
    Dropout(0.3),
    keras.layers.Dense(50,  kernel_initializer='lecun_normal', use_bias = False),
    keras.layers.BatchNormalization(),
    keras.layers.Activation('relu'),
    Dropout(0.1),
    Dense(total_noOf_classes, activation='softmax')
])
```

```python
cnn_model.compile(loss= keras.losses.SparseCategoricalCrossentropy(),
                  optimizer=keras.optimizers.Adam(learning_rate=0.01, beta_1 = 0.9, beta_2 = 0.999), 
                  metrics=['accuracy'])
trained_model = cnn_model.fit(X_train_rs, t_train,epochs=100, batch_size=16,validation_data=(X_val_rs,t_val),callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)])

numberofepochs = len(trained_model.history['loss'])
print("\nThe process of training the model fininshed with", numberofepochs, "epochs")
print('The accuracy obtained for the validation dataset is: %.2f%%'%(np.max(trained_model.history['val_accuracy'])*100))
```
